```python
# read in the data from the dictionary into pandas dataframe in one line of code
# set index to the id column 
df = pd.DataFrame.from_dict(data)
df = df.set_index('id')
df.head()
```

```python
# create a subdf where sub is a filtered version of the df for only one cuisine
df_indian = df[df.cuisine =='indian'] # indian
df_indian = df_indian.reset_index(drop = True)
df_indian
```

```python
# use collections to help figure out how many times a certain ingredient is used
from collections import Counter
```

```python
Counter(df_indian.ingredients[0])
```

```python
Counter(df.ingredients[0])
```

```python
#List of all ingredients in indian receipe: list of lists
all_indian_ings = list(df_indian.ingredients)
all_indian_ings[0:2] 
```

```python
# convert a list of lists to one list 
flat_indian_ings =[item for sublist in all_indian_ings for item in sublist]
```

```python
flat_indian_ings[0:2]
```

```python
indian_ings = Counter(flat_indian_ings)
indian_ings
```

```python
#indian ingredients df, converting counter dictionary to dataframe 
indian_ings_df = pd.DataFrame.from_dict(indian_ings, orient= 'index')
indian_ings_df.head(5)
```

```python
indian_ings_df.columns = ['count']
indian_ings_df.head()
```

```python
# top 10 ingredients in this sub cuisine? 
indian_ings_df = indian_ings_df.sort_values(by = 'count',ascending = False)
indian_ings_df[:10]
```

```python
len(indian_ings_df)
```

```python
in_allings_list = indian_ings_df.index
in_allings_list[0:10]
```

```python
# stop word search
stop_words_find = [w for w in in_allings_list if re.search(r'\b ?salt|pepper|sauce|water|seasoning|powder?\b', w)]
#stop_words_find
```

```python
# indian ingredients list wihtout stop words
in_ings_final = list(set(in_allings_list)-set(stop_words_find))
len(in_ings_final)
```

```python
in_df = pd.DataFrame(df_indian['ingredients'].to_list())
in_df.head() 
```

```python
in_tup_list = []
for i in in_ings_final:
    for j in range(len(all_indian_ings)):
        if i in all_indian_ings[j]:
            sub_tup_list = [(i, val) for val in all_indian_ings[j] if val not in stop_words_find and val != i]
            in_tup_list.append(sub_tup_list)
```

```python
# convert a list of lists to one list 
flat_in_tup_ings =[item for sublist in in_tup_list for item in sublist]
```

```python
flat_in_tup_ings[0:10]
```

```python
len(flat_in_tup_ings)
```

```python
in_ings_dict = Counter(flat_in_tup_ings)
#in_ings_dict
```

```python
in_ings_df1 = pd.DataFrame.from_dict(in_ings_dict, orient = 'index').reset_index()
in_ings_df1.head()
```

```python
in_ings_df1 = in_ings_df1.rename(columns = {0: 'Weight'})
in_ings_df1.head()
```

```python
Source, Target = zip(*in_ings_df1['index'])
```

```python
in_ings_df1['Source'] = pd.Series(Source)
in_ings_df1['Target'] = pd.Series(Target)
```
